## Language Analysis

```python
def detect_lang(text):
    """
    Helper function to detect the langauge of a given text using langid library
    params : 
        text (string) : the text to be detected
        
    return :
        lang : the detected language
    """
    try:
        lang, _ = langid.classify(text)
        return lang
    except Exception as e:
        print(f"Error detecting language", e)
        return None
```

```python
train_data['language'] = train_data['Description'].apply(detect_lang)
train_data.head()
```

```python
train_data['language'].value_counts()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x='language', data=train_data,palette='rainbow',edgecolor=(0,0,0))
plt.xlabel('Language')
plt.ylabel('Count')
plt.title('Visualize the Language Count')
plt.show()
```

We observe the presence of multiple languages in our training data. To tackle this, we will translate rows with different languages to English

```python
def trans(text):
    """
    Helper Function use google Translator API to translate a text to english
    params :
        text (string) : the text to be translated
    return :
        the translated text in the case of success the origin text otherwise
    """
    try:
        return Translator().translate(text, 'en').text
    except:
        return text
```

```python
train_data.loc[train_data['language'] != 'en','Description'] = train_data.loc[train_data['language'] != 'en','Description'].apply(trans)
```

## ðŸ”„ Data Preprocessing

```python
def preprocess_text(text):
    
    # Remove strange pattern in different languages if exist 
    text = re.sub('Mail <svaradi@sprynet.com> for translation. ','',text)
    # Remove twitter handles
    text = re.sub(r'@\S+', '', text)
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove punctuations
    text = re.sub(f'[{string.punctuation}]','',text)
    # Remove numbers
    text = re.sub(f'[{string.digits}]','',text)
    # Remove single charachters 
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)

    return text
```

```python
train_data['Description'] = train_data['Description'].apply(preprocess_text)
```

```python
train_data.head()
```

Let's visualize the most frequently used words in our training dataset. We'll be using the function below to generate a barplot containing the top n words

```python
def freq_words(x, n = 30): 
    """
    Helper Function
    Generate a bar plot to visualize the most frequently used n words in a text corpus.

    Parameters:
    - x (list): A list of text documents.
    - terms (int): Number of top words to display in the plot. Default is 30.

    Returns:
    None
    """
    all_words = ' '.join(text for text in x) 
    all_words = all_words.split() 
    fdist = nltk.FreqDist(all_words) 
    words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) 

    d = words_df.nlargest(columns="count", n = n) 

    plt.figure(figsize=(15,6)) 
    sns.barplot(data=d, x= "count", y = "word", palette='rainbow') 
    plt.show()

freq_words(train_data['Description'], 10)
```

We observe that the most frequently used words consist of stopwords, which hold no significant meaning for our movie classification task. To address this, we'll clean the data to eliminate stopwords. This way, we aim to distinguish between each genre based on words that carry more genre-specific relevance. We can do so by specifiying the params `stop_words` in the vectorizer instance.

```python
train_data.drop(columns='language', inplace=True)
```

```python
train_data
```

## ðŸ”§ Feature Extraction

We'll be using TF-IDF for feature extraction. Scikit-learn provides a predefined vectorizer for this purpose

```python
tfidf_vectorizer = TfidfVectorizer(lowercase=True,
                                   ngram_range=(1,1), 
                                   stop_words='english', # Eliminate Stop Words
                                   min_df=2)

x_train = tfidf_vectorizer.fit_transform(train_data['Description'])
```

To address the distribution imbalance issue, we'll be using oversampling. Imbalanced-learn **`imblearn`** comes with a predefined class for this purpose

```python
sampler = RandomOverSampler()

#We will pass to it the output of TfidfVectorizer from train data
x_train_resampled , y_train_resampled = sampler.fit_resample(x_train,train_data['Genre'])
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=y_train_resampled,x=y_train_resampled.values,palette='rainbow')
plt.xticks(rotation=90)
plt.show()
```

## ðŸš€ Train a Classifier

Time to train a Naive Bayes Classifier

```python
model = MultinomialNB(alpha=0.3)
model.fit(x_train_resampled, y_train_resampled)
```

## ðŸŽ¯ Make Predictions

```python
test_data = pd.read_csv(test_data_path, sep=':::', \
                         engine='python',names=['Title', 'Genre', 'Description']).reset_index(drop=True)
test_data.head()
```

```python
test_data['language'] = test_data['Description'].apply(detect_lang)
test_data.language.value_counts()
```

```python
test_data.loc[test_data['language'] != 'en','Description']
```

```python
test_data.loc[test_data['language'] != 'en','Description'] = test_data.loc[test_data['language'] != 'en','Description'].apply(trans)
test_data['Description'] = test_data['Description'].apply(preprocess_text)
```

```python
test_data.head()
```

```python
x_test = tfidf_vectorizer.transform(test_data['Description'])
```

```python
y_test = test_data['Genre'].to_numpy()
y_test
```

```python
y_pred = model.predict(x_test)
```

## ðŸ“ˆ Evaluate the Model

```python
def plot_cm(cm, classes):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
plot_cm(cm, model.classes_)
```

```python
print(accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

Looks like our Naive Bayes model is feeling a bit indecisive ðŸ—¿  
Let's try another classifiers
## Logistic regression

```python
clf = LogisticRegression(random_state=0, max_iter=1000).fit(x_train_resampled, y_train_resampled)
```

```python
y_pred = clf.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
plot_cm(cm, clf.classes_)
```

```python
print(accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

we can observe that logistic regression gives better accuracy compared to naive Bayes even though it is still low around 56%, this may be caused by the data distribution as we discussed earlier ðŸ˜¢
